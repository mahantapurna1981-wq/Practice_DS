```python
%pip install plotly
import plotly
import plotly.graph_objs as go
import plotly.express as px
import pandas as pd
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=[1, 2, 3,4,5], y=[1, 7,8,2,3], mode='markers'))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=[1, 2, 3,4,5], y=[1, 7,8,2,3], mode='lines'))
```

```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=tips.total_bill, y=tips.tip, mode='markers'))
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(x=[1,2,3,4,5], y=[3,8,5,2,6]))
```

```python
fig=go.Figure()
fig.add_trace(go.Histogram( x=tips.total_bill))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=tips.total_bill, y=tips.tip, mode='markers',marker_size=12*tips['size']))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter3d(x=tips.total_bill, y=tips.tip, mode='markers',marker_size=10*tips['size'], z= tips['size']))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter3d(x=[1,2,3,4,5], y=[3,8,5,2,6],z=[1,2,7,3,5]))
```

```python
# using plotly express
```

```python
x=[1,2,3,4]
y=[3,8,5,2]
```

```python
fig=px.scatter(x=x,y=y)
fig.show()
```

```python
fig=px.line(x=x,y=y)
fig.show()
```

```python
fig=px.bar(x=x,y=y)
fig.show()
```

```python
fig=px.box(x,y=[3,4,7,9,7,5,3])
fig.show()
```

```python
data=[[1,2,3],[4,5,6],[6,7,8]]
fig=px.imshow(data)
fig.show()
```

```python
x=[1,2,3,4]
y=[3,8,5,2]
fig=px.scatter(x,y,size=[30,50,10,200])
fig.show()
```

```python
x=[1,2,3,4]
y=[3,8,5,2]
z=[1,2,7,3]
fig=px.scatter_3d(x=x,y=y,z=z)
fig.show()
```
