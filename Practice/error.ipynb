```python
from bokeh.plotting import figure, show
from bokeh.sampledata.iris import flowers

# Initialize figure without default tools
p = figure(title='test_flower', x_axis_label='petal_length', y_axis_label='petal_width', tools='')

p.circle(flowers['petal_length'], flowers['petal_width'], size=6, color='blue', alpha=0.5)

show(p)
```
