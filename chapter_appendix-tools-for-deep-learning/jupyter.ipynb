# Restful API and Flassk

##**1. What is a RESTful API?**

**Answer:** A RESTful API is an application interface that follows REST
principles, using HTTP methods to perform operations on resources.

##**2. Explain the concept of API specification.**

**Answer:** An API specification defines how an API works, including
endpoints, request formats, response formats, authentication, and error
codes.

##**3. What is Flask, and why is it popular for building APIs?**

**Answer:** Flask is a lightweight Python web framework known for
simplicity, flexibility, and easy integration, making it ideal for
building APIs.

##**4. What is routing in Flask?**

**Answer:** Routing maps URL paths to specific functions in Flask,
determining what code runs when a user accesses a particular route.

##**5. How do you create a simple Flask application?**

**Answer:**

    from flask import Flask
    app = Flask(__name__)

    @app.route("/")
    def home():
        return "Hello World!"

    app.run()

##**6. What are HTTP methods used in RESTful APIs?**

**Answer:** Common methods are GET, POST, PUT, PATCH, DELETE, and OPTIONS.

##**7. What is the purpose of the @app.route() decorator in Flask?**

**Answer:** It binds a URL path to a function, turning it into an API
endpoint.

##**8. What is the difference between GET and POST HTTP methods?**

**Answer:**
- GET retrieves data.
- POST sends data to the server, often used for creating resources.

##**9. How do you handle errors in Flask APIs?**

##**10. How do you connect Flask to a SQL database?**

**Answer:** Using Flask-SQLAlchemy or other libraries to configure database
URIs and models.

##**11. What is the role of Flask-SQLAlchemy?**

**Answer:** It simplifies working with SQL databases by providing ORM
features.

##**12. What are Flask blueprints, and how are they useful?**

**Answer:** Blueprints allow modular application structure, grouping routes
into separate components.

##**13. What is the purpose of Flask’s request object?**

**Answer:** It gives access to incoming request data such as JSON, query
params, and form data.

##**15. What is the purpose of Flask’s jsonify() function?**

**Answer:** It converts Python dictionaries to proper JSON responses.


##**16. Explain Flask’s url_for() function.**

**Answer:** It generates URLs dynamically based on function names.

##**17. How does Flask handle static files (CSS, JavaScript, etc.)?

**Answer:** Flask automatically serves files placed inside the static/
folder.

##**18. What is an API specification, and how does it help in building a Flask API?**

**Answer:** It documents API endpoints, making development, testing, and
teamwork easier.

##**19. What are HTTP status codes, and why are they important in a Flask API?**

**Answer:** They indicate the result of API requests, such as 200 (success),
404 (not found), 500 (server error).

**Answer:** By using authentication (JWT, API keys), HTTPS, validation, and
